```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/Crop_recommendation.csv')
```

```python
df.head()
```

```python


print(df.info())
```

```python
X = df.drop("label", axis=1)

Y = df["label"]
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42
)
```

```python
X_train
```

```python
X_test
```

```python
model = DecisionTreeClassifier(
    random_state=42
)
```

```python
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
accuracy = accuracy_score(
    Y_test,
    predictions
)

print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(
    Y_test,
    predictions
)

print(cm)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.boxplot(x='label', y='rainfall', data=df)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))

plot_tree(
    model,
    feature_names=X.columns,
    class_names=model.classes_,
    filled=True
)

plt.show()
```

Why Decision Tree?

The target variable is the crop label (rice, maize, cotton, etc.), which is categorical. Decision Trees are suitable because they classify data using a series of IF-ELSE conditions based on rainfall, temperature, humidity, pH, and nutrient values. The model is easy to interpret and closely matches real-world agricultural decision-making.
